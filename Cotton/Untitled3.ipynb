```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
import urllib.request

html = urlopen('https://www.shutterstock.com/search/cotton+fields?page=6')
bs = BeautifulSoup(html, 'html.parser')
images = bs.find_all('img', {'src':re.compile('.jpg')})
i=621
for image in images: 
    print(image['src']+'\n')
    i=i+1
    urllib.request.urlretrieve(image['src'], "cotton"+str(i)+".jpg")
```
