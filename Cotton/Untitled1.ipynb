```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
import urllib.request

html = urlopen('https://www.shutterstock.com/search/cotton+crop')
bs = BeautifulSoup(html, 'html.parser')
images = bs.find_all('img', {'src':re.compile('.jpg')})
i=0
for image in images: 
    print(image['src']+'\n')
    i=i+1
    urllib.request.urlretrieve(image['src'], "cotton"+str(i)+".jpg")
```
